```python
def metalstore():                                                                                    # --- defining a function --- #
    cart= {}
    total= 0
    intro= ''                                                                
    while intro != 0:
        print('Welcome to David"s Precious Metals. This is the main menu')                           # --------- main menu --------#
        intro= int(input('Enter 1 to begin, 2 to view cart or 0 to exit '))
        if intro == 1:                                                                               # - Menu option 1 (shop) Begin - #
            print('We have 3 item categories: Gold , Silver, and Copper. Which of these would you like? ') 
            cg= int(input('1 for Gold, 2 for Silver, 3 for Copper '))                                #           sub-menu
            if cg == 1:                                                                              # ----- sub category 1  -----#
                print('We have 1oz Coins, 5oz Bars and 10oz Bars. Which of these would you like?')  
                item_select= int(input('Enter 1 for 1oz Coin, 2 for 5oz Bar, 3 for 10oz Bar '))      # item select
                quantity= int(input('Enter item quantity'))                                          # ---- enters quantity ---- #
                if item_select == 1:
                    cart['1oz Gold Coin'] = quantity
                    total += quantity * 2000
                if item_select == 2:                                                                        #items 1-3 being added to cart
                    cart['5oz Gold Bar'] = quantity
                    total += quantity * 10000
                if item_select == 3:                                    
                    cart['10oz Gold Bar'] = quantity
                    total += quantity * 20000
            elif cg == 2:
                print('We have 1oz Coins, 5oz Bars and 10oz Bars. Which of these would you like?')           # sub category 2                   
                item_select= int(input('Enter 1 for 1oz Coin, 2 for 5oz Bar, 3 for 10oz Bar '))             #  item select
                quantity= int(input('Enter item quantity '))                                            # ------ enters quantity ------#
                if item_select == 1:
                    cart['1oz Silver Coin'] = quantity
                    total += quantity * 30
                if item_select == 2:
                    cart['5oz Silver Bar'] = quantity                                                      #4-6 etc
                    total += quantity * 150
                if item_select == 3:
                    cart['10oz Silver Bar'] = quantity
                    total += quantity * 300
            elif cg == 3:
                print('We have 1oz Coins, 5oz Bars and 10oz Bars. Which of these would you like?')            #sub category 3
                item_select= int(input('Enter 1 for 1oz Coin, 2 for 5oz Bar, 3 for 10oz Bar '))               # item select
                quantity= int(input('Enter item quantity '))                                              #  - -- enters quantity ----  #
                if item_select == 1:
                    cart['1oz Copper Coin'] = quantity
                    total += quantity * 2
                if item_select == 2:
                    cart['5oz Copper Bar'] = quantity
                    total += quantity * 10
                if item_select == 3:
                    cart['10oz Copper Bar'] = quantity
                    total += quantity * 20
            else:
                print('invalid entry')                                                                  # ------sub category 3 end----- #
        elif intro == 2:
            print(cart, 'Your total comes to $', total)
            ask= int(input('What would you like to do? 1 to checkout or 2 to remove items '))
            if ask == 2:                                                                                 # menu option 2.2 (view cart) begins
                print('Items are numbered as follows: 1oz Gold Coin= 1, 5oz Gold Bar= 2, 10oz Gold Bar=3. Silver= 4-6, etc.')
                ask2= int(input('Select the item you"d like removed by entering 1-9 or 0 to remove them all '))
                if ask2 == 1:                                                                         # this is option to remove 1 of the 9 possible items in cart 
                    if '1oz Gold Coin' in cart:
                        total -= cart['1oz Gold Coin'] * 2000
                        cart.pop('1oz Gold Coin')
                        print('All 1oz Gold Coins have been removed')                                      #1
                    else: print('Item is not in cart.')
                elif ask2 == 2:
                    if '5oz Gold Bar' in cart:
                        total -= cart['5oz Gold Bar'] * 10000
                        cart.pop('5oz Gold Bar')
                        print('All 5oz Gold Bars have been removed')                                            #2
                    else: print('Item is not in cart.')
                elif ask2 == 3:
                    if '10oz Gold Bar' in cart:
                        total -= cart['10oz Gold Bar'] * 20000
                        cart.pop('10oz Gold Bar')
                        print('All 10oz Gold Bars have been removed')                                                #3 and so on. 
                    else: print('Item is not in cart.')
                elif ask2 == 4:
                    if '1oz Silver Coin' in cart:
                        total -= cart['1oz Silver Coin'] * 30
                        cart.pop('1oz Silver Coin')
                        print('All 1oz Silver Coins have been removed')
                    else: print('Item is not in cart.')
                elif ask2 == 5:
                    if '5oz Silver Bar' in cart:
                        total -= cart['5oz Silver Bar'] * 150
                        cart.pop('5oz Silver Bar')
                        print('All 5oz Silver Bars have been removed')
                    else: print('Item is not in cart.')
                elif ask2 == 6:
                    if '10oz Silver Bar' in cart:
                        total -= cart['10oz Silver Bar'] * 300
                        cart.pop('10oz Silver Bar')
                        print('All 10oz Silver Bars have been removed')
                    else: print('Item is not in cart.')
                elif ask2 == 7:
                    if '1oz Copper Coin' in cart:
                        total -= cart['1oz Copper Coin'] * 2
                        cart.pop('1oz Copper Coin')
                        print('All 1oz Copper Coins have been removed')
                    else: print('Item is not in cart.')
                elif ask2 == 8:
                    if '5oz Copper Bar' in cart:
                        total -= cart['5oz Copper Bar'] * 10
                        cart.pop('5oz Copper Bar')
                        print('All 5oz Copper Bars have been removed')
                    else: print('Item is not in cart.')
                elif ask2 == 9:
                    if '10oz Copper Bar' in cart:
                        total -= cart['10oz Copper Bar'] * 20
                        cart.pop('10oz Copper Bar')
                        print('All 10oz Copper Bars have been removed')
                    else: print('Item is not in cart.')
                elif ask2 == 0:
                    cart.clear()
                    print('Your cart is now empty')
                else:
                    print('Invalid entry')
            else:
                print(cart, 'Your total comes to $', total)
                break
        else:
            cart.clear()
            print('Your cart has been cleared. Goodbye')
            break
```

```python
metalstore()
```
